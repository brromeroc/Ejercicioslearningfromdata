**(a)** Si $w^*$ es el optimo del conjunto de pesoso que separa la informaci√≥n, entonces para todo $x_n$, en el Data set $D$, se tiene que $w^*x_n$ tiene el mismo signos que $y_n$, lo cual siginifica que $y_n(w^{*^{T}}x_n)>0$, esto implica que $œÅ=min_{1\leq n\leq N} y_n(w^{*^{T}}x_n)>0$

**(b)** Provemos esta desigualdad utilizando inducci√≥n sobre $t$, cuando $t=1$ tenemos que  

$$w^T(t-1)w^* + \rho = w^T(0)w^* + \rho = \rho$$

Mientras que 

$$w(t)=w(1) = w(0) + y_ix_i = y_ix_i$$

Para algun $i$, Asi  $w^T(t)w^* = (y_ix_i)^Tw^* = y_ix_i^Tw^* = y_iw^{*T}x_i$, donde utilizamos el hecho de que $y_i\in \{-1,1\}$ y $x_i^Tw^*$ es un escalar. Por  definci√≥n de $\rho$, tenemos que $y_iw^{*T} x_i \ge \rho$, de esta manera esta afimaci√≥n es cierta para $t=1$.
Ahora suponga que la afirmaci√≥n es valida para $t$ vamos a provar que sigue siendo valida para $t+1$ y $w(t+1)=w(t)+y_ix_i$, para alg√∫n $i$, entonces 

$$\begin{align*}
w^T (t+1)w^* &= w^T(t)w^* + y_ix_i^Tw^* \\
&\ge w^T(t-1)w^* + \rho + y_ix_i^Tw^*  \\
&= \left(w^T(t-1) + y_ix_i^T\right)w^* + \rho \\
&= w^T(t)w^* + \rho\\
\end{align*}$$

Esto finaliza la prueba, adem√°s podemos ver que 

$w^T(t)w^* \ge w^T(t-1)w^* + \rho \ge w^T(t-2)w^* + 2\rho \ge \dots \ge t\rho$


**(c)** $w(t) = w(t-1) + y(t-1)x(t-1)$, ahora tome el cuadrado de la norma de $w(t)$, tenemos que:

$$\begin{align*}
||w(t)||^2 &= \left(w(t-1) + y(t-1)x(t-1)\right)^T\left(w(t-1) + y(t-1)x(t-1)\right)\\
&= ||w(t-1)||^2 + ||y(t-1)||^2||x(t-1)||^2 + 2y(t-1)w^T(t-1)x(t-1) \\
&= ||w(t-1)||^2 + ||x(t-1)||^2 + 2y(t-1)w^T(t-1)x(t-1) \\
&\le ||w(t-1)||^2 + ||x(t-1)||^2
\end{align*}$$

dado que $y(t-1)w^T(t-1)x(t-1)\leq 0$, ya que $x(t-1)$ fue clasificada erroneamente por $w(t-1)$


**(d)** Utilicmos inducci√≥n para provar lo anterior 

Cuando $t=0$ es claro que $||w(0)||^2=0\leq tR^2=0$, ahora asuma que que se tiene para $t$, es decir $$||w(t)||^2\leq tR^2$$,. entonces para $t+1$ tenemos que $||w(t)||^2 \le ||w(t-1)||^2 + ||x(t-1)||^2 \le (t-1)R^2 + R^2 = tR^2$. por el anterior problema tenemos que $$||w(t)||^2 \le ||w(t-1)||^2 + ||x(t-1)||^2 \le (t-1)R^2 + R^2 = tR^2$$

Aqui usamos el hecho que $||x(t-1)||^2\leq R^2$, por la definic√≥n de $R$


**(e)** Aplicando los resultados de $b$ y $d$, obtenemso que 

$$\frac{w^T(t)}{||w(t)||} w^* \ge \frac{t\rho}{||w(t)||} \ge  \frac{t\rho}{\sqrt{t}R} = \sqrt{t}\frac{\rho}{R}$$

Invirtiendo la inecuaci√≥n y tomando el cuadrado tenemos que
